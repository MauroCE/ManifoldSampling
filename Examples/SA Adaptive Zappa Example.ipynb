# Imports and Settings

```python
import sys
sys.path.append("..")
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from Zappa.zappa import zappa_adaptive, zappa_sampling
from utils import logf_Jacobian
from utils import logp as logp_scale
from utils import quick_MVN_marginals, quick_MVN_scatter, update_scale_sa
from HMC.gaussian_hmc import GaussianTargetHMC
from Manifolds.RotatedEllipse import RotatedEllipse
```

```python
# Settings Parameters
n = 1000
s = 0.5
tol = 1.48e-08
a_guess = 1.0
ap_star = 0.6

# Target Distribution
mu = np.zeros(2)
Sigma = np.array([[1.0, 0.99], [0.99, 2.0]]) 
target = multivariate_normal(mean=mu, cov=Sigma)

# Initial Point & Contour
incr = 0.001
x0 = mu + incr 
z = target.pdf(x0)
ellipse = RotatedEllipse(mu, Sigma, z)

# Target & proposal on manifold
logf = lambda xy: logf_Jacobian(xy, Sigma)
logp = lambda xy: logp_scale(xy, s)

# Plotting parameters
xlims = (-incr*5, incr*5)
ylims = (-incr*5, incr*5)
step = 0.00001

```

# Let's Compare the standard Zappa Algorithm with Adaptive Zappa

$$
\lambda_{n+1} = \lambda_n + \frac{1}{\sqrt[3]{n^2}}(\alpha_{\lambda(u)}(x, u) - \alpha_*)
$$

```python
samples_adaptive = zappa_adaptive(x0, ellipse, logf, n, s, tol, a_guess, ap_star, update_scale_sa)
```

```python
samples = zappa_sampling(x0, ellipse, logf, logp, n, s, tol, a_guess)
```

```python
quick_MVN_scatter(samples_adaptive, target, xlims=xlims, ylims=ylims, levels=[z], step=step, lw=1)
```

```python
quick_MVN_scatter(samples, target, xlims=xlims, ylims=ylims, levels=[z], step=step, lw=1)
```

# Analysis of log-scale and log-ap behavior

```python
ZappaObj = ZappaAdaptive(x0, ellipse, logf, logp_scale, s, n, tol, a_guess, lambda x: x, ap_star, 2/3)
out = ZappaObj.sample()
```

```python
print("Final Scale: {s} \nAcceptance Rate: {ar}".format(s=ZappaObj.s, ar=ZappaObj.acceptance_rate()))
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(20, 5))
ax[0].plot(out['logaps'])
ax[0].set_title(r"$\log \alpha_{\lambda(u)}(x, u)$", fontsize=20)
ax[1].plot(out['LogScales'])
ax[1].set_title(r"$\lambda_n(u)$", fontsize=20)
plt.show()
```
