```python
# True Marginals
marginal_x = lambda x: scipy.stats.norm(loc=target.mean[0], scale=np.sqrt(target.cov[0, 0])).pdf(x)
marginal_y = lambda y: scipy.stats.norm(loc=target.mean[1], scale=np.sqrt(target.cov[1, 1])).pdf(y)
```

# Hug & Hop

```python
# HUG AND HOP
fig, ax = plt.subplots(ncols=2, figsize=(20, 5))
ax[0].plot(x, marginal_x(x), label="True Marginal")
ax[0].plot(x, xkde_hh, label="HH")
ax[0].plot(x, xkde_hhpc, label="HHPC")
ax[0].set_xlabel("X Marginal", fontsize=20)
ax[1].plot(x, marginal_y(x), label="True Marginal")
ax[1].plot(x, ykde_hh, label="HH")
ax[1].plot(x, ykde_hhpc, label="HHPC")
ax[1].set_xlabel("Y Marginal", fontsize=20)
plt.legend()
fig.suptitle("Huh and Hop", fontsize=20)
plt.show()
```
