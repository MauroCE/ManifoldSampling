# Hug
Hug decouples the proposal mechanism from the acceptance mechanism to target a distribution $\pi$ around $f^{-1}(0)$ for $f:\mathbb{R}^n\to\mathbb{R}$
#### Proposal Mechanism
The proposal mechanism constructs a fictitious constrained Hamiltonian $H_p$
$$
H_{p} = \frac{1}{2}v^\top v + f(x) \lambda
$$
with constant potential function and quadratic kinetic energy, it then solves for the Lagrange multiplier
$$
\lambda(x, v) = \frac{v^\top \nabla_x^2 f(x) v}{\|\nabla_x f(x)\|^2}
$$
and constructs the following Constrained Hamiltonian System
$$
\begin{align}
    \dot{x} &= v \\
    \dot{v} &= - \frac{v^\top \nabla_x^2 f(x) v}{\|\nabla_x f(x)\|}\widehat{\nabla_x f}(x).
\end{align}
$$
This is a non-separable system and Hug integrates this using an approximate GPV
$$
\begin{align}
    x_{n + 1/2} &= x_n + \frac{\delta}{2}v_n \\ 
    v_{n + 1}   &= v_n - \delta \frac{v_n^\top H(x_{n+1/2}) v_n}{\|\nabla f(x_{n+1/2})\|}\widehat{\nabla f}(x_{n+1/2}) \\
    x_{n + 1}   &= x_{n+1/2} + \frac{\delta}{2} v_{n+1}
\end{align}
$$
This is explicit, non-reversible and non-symplectic, but using the Hug approximation of the Hessian we obtain
$$
\begin{align}
    x_{n + 1/2} &= x_n + \frac{\delta}{2}v_n \\ 
    v_{n + 1}   &= v_n - 2\widehat{\nabla f}(x_{n+1/2}) \widehat{\nabla f}(x_{n+1/2})^\top v_n\\
    x_{n + 1}   &= x_{n+1/2} + \frac{\delta}{2} v_{n+1}
\end{align}
$$
which is now reversible and volume-preserving. 

#### Acceptance Mechanism
The acceptance mechanism uses a different Hamiltonian 
$$
H_a(x, v) = -\log \widetilde{p}_\pi(x) + \frac{1}{2}v^\top v
$$
hence the acceptance function is 
$$
a(x, v, x', v') = \min\left\{1, e^{- H_a(x', v') + H_a(x, v)}\right\} = \min\left\{1, \exp\left\{\log\widetilde{p}_\pi(x') - \log\widetilde{p}_\pi(x) - \|v'\|^2/2 + \|v\|^2/2\right\}\right\}
$$
Overall we describe the algorithm as $(H_p, \Phi_{\text{AGPV}}, H_a)$ to say $H_p$ is the Hamiltonian used for the proposal mechanism, $\Phi_{\text{AGPV}}$ is the integrator used to discretize $H_p$, and $H_a$ is used for the acceptance function.

# Alternative

Alternatively, one could simply replace the integrator used. That is, one could use $(H_p, \Phi, H_a)$ for a different integrator. For instance, we have at hand a Partitioned System
$$
\begin{align}
    \dot{x} &= F_1(x, v) \\
    \dot{v} &= F_2(x, v)
\end{align}
$$
where in our case
$$
\begin{align}
    F_1(x, v) &= v \\ 
    F_2(x, v) &= - \frac{v^\top \nabla_x^2 f(x) v}{\|\nabla_x f(x)\|}\widehat{\nabla_x f}(x)
\end{align}
$$
and it is possible to use the stormer-verlet for partitioned systems. In particular the Generalized Position Verlet
$$
\begin{align}
    x_{n+1/2} &= x_n + \frac{\delta}{2} F_1(x_{n + 1/2}, v_{n}) \\
    v_{n + 1} &= v_n + \frac{\delta}{2}(F_2(x_{n+1/2}, v_n) + F_2(x_{n+1/2}, v_{n+1})) \\
    x_{n + 1} &= x_{n+1/2} + \frac{\delta}{2} F_1(x_{n+1/2}, v_{n+1})
\end{align}
$$
This is implicit, second-order, symmetric and symplectic. In our case this becomes

$$
\begin{align}
    x_{n+1/2} &= x_n + \frac{\delta}{2} v_n \\
    v_{n + 1} &= v_n + \frac{\delta}{2}\left[- \frac{v_n^\top H_{n+1/2} v_n}{\|\nabla f(x_{n+1/2})\|}\widehat{\nabla f}(x_{n+1/2}) - \frac{v_{n+1}^\top H_{n+1/2} v_{n+1}}{\|\nabla f(x_{n+1/2})\|}\widehat{\nabla f}(x_{n+1/2})\right] \\
    x_{n + 1} &= x_{n+1/2} + \frac{\delta}{2} v_{n+1}
\end{align}
$$

The velocity update is implicit, but actually can be solved explicitly. Indeed denote the following variables
$$
\begin{align}
    a &= v_n -\frac{\delta}{2}\frac{v_n^\top H_{n+1/2} v_n}{\|\nabla f(x_{n+1/2})\|}\widehat{\nabla f}(x_{n+1/2}) \\
    b &= -\frac{\delta}{2}\frac{\widehat{\nabla f}(x_{n+1/2})}{\|\nabla f(x_{n+1/2})\|}
\end{align}
$$
where $a\in\mathbb{R}^n$ and $b\in\mathbb{R}^n$. Of course we have the relationship
$$
a = v_n + b v_n^\top H_{n+1/2} v_n
$$
and the velocity update becomes
$$
v_{n+1} = a + b v_{n+1}^\top H_{n+1/2} v_{n+1}
$$

The last expression means that there must exist a $\gamma\geq 0$ (since Hessian is positive semi-definite) such that
$$
v_{n+1} = a + b\gamma
$$
Substituting this into the expression gives
$$
a + b\gamma = a + b(a + b\gamma)^\top H_{n+1/2}(a + b\gamma)
$$
This simplifies to (writing $H = H_{n+1/2}$ and remembering $H$ is symmetric)
$$
b\gamma = b\left[a^\top H a + 2\gamma a^\top H b + \gamma^2 b^\top H b\right]
$$

Again, this implies
$$
\gamma^2 b^\top H b + \gamma(2a^\top H b - 1) + a^\top H a = 0
$$

The quadratic formula gives
$$
\gamma_{1, 2} = \frac{-2 a^\top H b + 1 \pm \sqrt{(2 a^\top H b - 1)^2 - 4 (b^\top H b) (a^\top H a)}}{2b^\top H b}
$$

```python
def integratorγ1(ξ):
    """This integrator always chooses γ1."""
    x0, v0 = ξ[:2], ξ[2:]
    x1 = x0 + (δ/2)*v0
    g1 = grad_f(x1); gnorm1 = norm(g1); g1hat = g1 / gnorm1
    H1 = hess_f(x1)
    b = -(δ/2)* g1hat / gnorm1
    a = v0 + b * (v0 @ (H1 @ v0))
    c1 = b @ (H1 @ b)
    c2 = 2 * a @ (H1 @ b) - 1
    c3 = a @ (H1 @ a)
    Δ = c2**2 - 4*c1*c3
    if Δ >= 0:
        γ = (-c2 + sqrt(Δ)) / (2*c1)
    else:
        print("Discriminat is negative.")
        return None
    v1 = a + b*γ
    x2 = x1 + (δ/2) * v1
    return np.concatenate((x2, v1))

def integratorγ2(ξ):
    """This integrator always chooses γ1."""
    x0, v0 = ξ[:2], ξ[2:]
    x1 = x0 + (δ/2)*v0
    g1 = grad_f(x1); gnorm1 = norm(g1); g1hat = g1 / gnorm1
    H1 = hess_f(x1)
    b = -(δ/2)* g1hat / gnorm1
    a = v0 + b * (v0 @ (H1 @ v0))
    c1 = b @ (H1 @ b)
    c2 = 2 * a @ (H1 @ b) - 1
    c3 = a @ (H1 @ a)
    Δ = c2**2 - 4*c1*c3
    if Δ >= 0:
        γ = (-c2 - sqrt(Δ)) / (2*c1)
    else:
        print("Discriminat is negative.")
        return None
    v1 = a + b*γ
    x2 = x1 + (δ/2) * v1
    return np.concatenate((x2, v1))

def integratorγ2_withδ(ξ, δ):
    """This integrator always chooses γ1."""
    x0, v0 = ξ[:2], ξ[2:]
    x1 = x0 + (δ/2)*v0
    g1 = grad_f(x1); gnorm1 = norm(g1); g1hat = g1 / gnorm1
    H1 = hess_f(x1)
    b = -(δ/2)* g1hat / gnorm1
    a = v0 + b * (v0 @ (H1 @ v0))
    c1 = b @ (H1 @ b)
    c2 = 2 * a @ (H1 @ b) - 1
    c3 = a @ (H1 @ a)
    Δ = c2**2 - 4*c1*c3
    if Δ >= 0:
        γ = (-c2 - sqrt(Δ)) / (2*c1)
    else:
        raise ValueError("Negative Discriminant.")
    v1 = a + b*γ
    x2 = x1 + (δ/2) * v1
    return np.concatenate((x2, v1))

def integratorγ2_many(ξ):
    """This integrator always chooses γ1."""
    x, v = ξ[:2], ξ[2:]
    
    for _ in range(B):
        x = x + (δ/2)*v
        g1 = grad_f(x); gnorm1 = norm(g1); g1hat = g1 / gnorm1
        H1 = hess_f(x)
        b = -(δ/2)* g1hat / gnorm1
        a = v + b * (v @ (H1 @ v))
        c1 = b @ (H1 @ b)
        c2 = 2 * a @ (H1 @ b) - 1
        c3 = a @ (H1 @ a)
        Δ = c2**2 - 4*c1*c3
        if Δ >= 0:
            γ = (-c2 - sqrt(Δ)) / (2*c1)
        else:
            print("Discriminat is negative.")
            break
        v = a + b*γ
        x = x + (δ/2) * v
    return np.concatenate((x, v))

def true_hug(x0, T, B, N, q, logpi, grad_f, hess_f):
    samples, acceptances = x0, zeros(N)
    for i in range(N):
        v0 = q.rvs()
        v, x = v0, x0
        logu = log(rand())
        δ = T / B
        for _ in range(B):
            x = x + (δ/2)*v
            g1 = grad_f(x); gnorm1 = norm(g1); g1hat = g1 / gnorm1
            H1 = hess_f(x)
            b = -(δ/2)* g1hat / gnorm1
            a = v + b * (v @ (H1 @ v))
            c1 = b @ (H1 @ b)
            c2 = 2 * a @ (H1 @ b) - 1
            c3 = a @ (H1 @ a)
            Δ = c2**2 - 4*c1*c3
            if Δ >= 0:
                γ = (-c2 - sqrt(Δ)) / (2*c1)
            else:
                print("Discriminat is negative.")
                break
            v = a + b*γ
            x = x + (δ/2) * v
        if logu <= logpi(x) + q.logpdf(v) - logpi(x0) - q.logpdf(v0):
            samples = np.vstack((samples, x))
            acceptances[i] = 1         # Accepted!
            x0 = x
        else:
            samples = np.vstack((samples, x0))
            acceptances[i] = 0         # Rejected
    return samples[1:], acceptances


def true_thug(x0, T, B, N, α, q, logpi, grad_f, hess_f):
    samples, acceptances = x0, zeros(N)
    for i in range(N):
        v0 = q.rvs()
        g0 = grad_f(x0); g0hat = g0 / norm(g0)
        w0 = v0 - α * g0hat * (v0 @ g0hat)
        w, x = w0, x0
        logu = log(rand())
        δ = T / B
        for _ in range(B):
            x = x + (δ/2)*w
            g1 = grad_f(x); gnorm1 = norm(g1); g1hat = g1 / gnorm1
            H1 = hess_f(x)
            b = -(δ/2)* g1hat / gnorm1
            a = w + b * (w @ (H1 @ w))
            c1 = b @ (H1 @ b)
            c2 = 2 * a @ (H1 @ b) - 1
            c3 = a @ (H1 @ a)
            Δ = c2**2 - 4*c1*c3
            if Δ >= 0:
                γ = (-c2 - sqrt(Δ)) / (2*c1)
            else:
                print("Discriminat is negative.")
                break
            w = a + b*γ
            x = x + (δ/2) * w
        g = grad_f(x) / norm(grad_f(x))
        v = w + (α / (1 - α)) * g * (g @ w)
        if logu <= logpi(x) + q.logpdf(v) - logpi(x0) - q.logpdf(v0):
            samples = np.vstack((samples, x))
            acceptances[i] = 1         # Accepted!
            x0 = x
        else:
            samples = np.vstack((samples, x0))
            acceptances[i] = 0         # Rejected
    return samples[1:], acceptances
```

```python
def check_order(x0, N, δ, perpendicular=False, thug=False, α=None):
    """Checks if true hug is second order."""
    average_difference = 0.0
    i = 0
    while i <= N:
        v0 = q.rvs()
        if perpendicular:
            v0norm = norm(v0)
            g0 = grad_f(x0); g0hat = g0 / norm(g0)
            v0 = v0 - (v0 @ g0hat) * g0hat
            v0 = v0 * v0norm / norm(v0)
        elif not perpendicular and thug and α is not None:
            g0 = grad_f(x0); g0hat = g0 / norm(g0)
            v0 = v0 - α * (v0 @ g0hat) * g0hat
        ξ0 = np.concatenate((x0, v0))
        try:
            ξ1 = integratorγ2_withδ(ξ0, δ)
        except ValueError:
            continue
        else:
            average_difference += abs(f(ξ1[:2]) - f(ξ0[:2])) 
            i += 1
    return average_difference / i
```
