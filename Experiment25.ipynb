# Box Plots for $T=0.01$

```python
columns = EPSILONS
width = 0.21
thing_that_changes = ALPHAS
n_alphas = len(thing_that_changes)
ticks = arange(start=1, stop=len(columns)*2, step=2)
n_boxes = n_alphas + 1
disp = linspace(-width * (n_boxes-1) / 2, width * (n_boxes-1) / 2, n_boxes, endpoint=True)
positions = disp.reshape(-1, 1) + ticks
colors1 = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'black']
colors2 = ['tan', 'cyan', 'lightgreen', 'yellow', 'violet', 'orange', 'lightgray']
labels = ['HUG'] + ['RHUG $\\alpha$={}'.format(thing) for thing in thing_that_changes]


index = -1 # T = 0.01

ARRAYS = [
    (THETA_ESS_HUG[:,:,index], THETA_ESS_RHUG[:,:,index,:]),
    (U_ESS_HUG[:,:,index], U_ESS_RHUG[:,:,index,:]),
    (A_HUG[:,:,index], A_RHUG[:,:,index,:]),
    (ESS_HUG[:,:,index], ESS_RHUG[:,:,index,:]),
    (EJSD_HUG[:,:,index], EJSD_RHUG[:,:,index,:])
]
YLABELS = [
    'ESS $\\theta$', 
    'ESS $u$',
    'AP',
    'multiESS',
    'EJSD'
]
n_plots = len(ARRAYS)
fig, ax = plt.subplots(nrows=n_plots, figsize=(20, 8 * n_plots))
for ix, (array1, array2) in enumerate(ARRAYS):
    bp1 = box_plot(ax[ix], array1.T.tolist(), 'red', 'tan', positions[0])
    bps = []
    for i in range(n_alphas):
        bps.append(box_plot(ax[ix], array2[:,:,i].T.tolist(), colors1[i+1], colors2[i+1], positions[i+1]))
    handles = [bp1["boxes"][0]] + [bp["boxes"][0] for bp in bps]
    ax[ix].legend(
        handles, 
        labels, 
        loc='lower left')
    ax[ix].set_xticks(ticks)
    ax[ix].set_xticklabels(columns)
    ax[ix].set_xlabel("$\\epsilon$", fontsize=20)
    ax[ix].set_ylabel(YLABELS[ix], fontsize=20)
    ax[ix].semilogy()

plt.show()
```

Suppose $\theta = (a, b, g, k)$ and $u = z$. Then the G&K simulator can be thought of as the following deterministic function $f:\mathbb{R}^6\to\mathbb{R}^m$ where $m$ is the number of simulations, in our case $m=2$. For this reason, we expect the Jacobian to be a matrix $\mathbb{R}^{2\times 6}$. 

Suppose we have observed $y^*\in\mathbb{R}^2$. Recall that the manifold will actually be the distance manifold, that is
$$
\mathcal{M}_\epsilon = \left\{\xi\in\mathbb{R}^6\,:\, \|f(\xi) - y^*\|_2 = \epsilon \right\}
$$
with the manifold for the true observation being
$$
\mathcal{M}_0 = \left\{\xi\in\mathbb{R}^6\,:\, \|f(\xi) - y^*\|_2 = 0 \right\} = \left\{\xi\in\mathbb{R}^5\,:\, f(\xi) = y^*\right\}.
$$

For this reason, the Jacobian is actually given by 
$$
J = \frac{J_f^\top (f(\xi) - y^*)}{\epsilon} \in\mathbb{R}^6
$$

The $i^{\text{th}}$ entry of the G&K simulator gives

$$
f_i(\xi) = a + b\left[1 + 0.8\frac{1 - \exp(-gz_i)}{1 + \exp(-gz_i)}\right](1 - z_i)^k z_i
$$

The derivatives with respect to the parameters are 

$$
\begin{align}
\nabla_a f_i(\xi) &= 1 \\
\nabla_b f_i(\xi) &= \left[1 + 0.8\frac{1 - \exp(-gz_i)}{1 + \exp(-gz_i)}\right](1 - z_i)^k z_i \\
\nabla_g f_i(\xi) &= 1.6 b (1 - z_i)^k z_i^2 \frac{e^{-gz_i}}{(1 + e^{-gz_i})^2} \\
\nabla_k f_i(\xi) &= b\left[1 + 0.8\frac{1 - \exp(-gz_i)}{1 + \exp(-gz_i)}\right](1 - z_i)^k z_i \ln(1 - z_i)
\end{align}
$$
