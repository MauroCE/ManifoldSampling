#### Target

```python
# Target is a MVN with mean `mu` and covariance `Sigma`
mu = np.zeros(2)                        
Sigma = np.array([[1.0, 0.99], [0.99, 2.0]])
target = multivariate_normal(mu, Sigma)

# Functions for log pdf and **normalized** gradient of log pdf
logpi = target.logpdf
grad_log_pi = lambda xy: - solve(Sigma, xy)
def norm_grad_log_pi(xy):
    gradient = grad_log_pi(xy)
    return gradient / norm(gradient)

# Proposal distribution is a standard MVN
q = multivariate_normal(np.zeros(2), np.eye(2))
```

#### HUG

```python
def Hug(x0, delta, q, logpi, grad_log_pi):
    v0 = q.rvs()
    v, x = v0, x0
    logu = np.log(rand())
    # Move
    x = x + delta*v/2 
    # Reflect
    g = grad_log_pi(x)
    ghat = g / norm(g)
    v = v - 2*(v @ ghat) * ghat
    # Move
    x = x + delta*v/2
    return x
```

```python
# Initial position and spherical velocity
x0 = target.rvs()
v0s = q.rvs()

# Settings
M = 50                                     # Number of times to average each run over
deltas = np.array([1.0, 0.1, 0.01, 0.001, 0.0001]) # Step size

# Store absolute value of difference of logq. Rows are different deltas, columns are different alphas.
potential_diff = np.zeros(len(deltas))

for ix, delta in enumerate(deltas):
    avg = 0.0
    for i in range(M):
        x_final = Hug(x0, delta, q, logpi, norm_grad_log_pi)
        avg += np.abs(logpi(x_final) - logpi(x0)) / M
    potential_diff[ix] = avg
```

```python
# Fit a linear regression model to find the slope
model = LinearRegression().fit(np.log10(deltas).reshape(-1, 1), np.log10(potential_diff))
model.coef_, model.intercept_
```

```python
logspace = np.logspace(-4, 0, base=10)

fig, ax = plt.subplots()
lines = ax.plot(deltas, potential_diff, marker='.', lw=3)
ax.set_xlabel(r'Log Scale $\delta$', fontsize=20)
ax.set_ylabel("Log Scale " + r'$|\ell(x_1) - \ell(x_0)|$', fontsize=12)
ax.set_xscale('log')
ax.set_yscale('log')
ax.plot(logspace, (10**model.intercept_)*(logspace**model.coef_), 
        color='orange', label='Fitted Line - Slope {:.3}'.format(model.coef_[0]))
ax.set_title("Change in Potential Energy - HUG", fontsize=20)
plt.legend()
plt.show()
```

#### THUG

```python
def THUG(x0, v0s, delta, alpha, q, logpi, norm_grad_log_pi):
    g = norm_grad_log_pi(x0)         # Compute normalized gradient at x0
    v0 = v0s - alpha * g * (g @ v0s) # Tilt velocity
    v, x = v0, x0                    # Housekeeping

    # B = 1 to check what happens in 1 step
    x = x + delta*v/2           # Move to midpoint
    g = norm_grad_log_pi(x)     # Compute gradient at midpoint
    v = v - 2*(v @ g) * g       # Reflect velocity using midpoint gradient
    x = x + delta*v/2           # Move from midpoint to end-point
    return x
```

```python
# Settings
M = 50                                     # Number of times to average each run over
alphas = np.array([0.0, 0.3, 0.6, 0.9])    # Squeezing factors
deltas = np.array([1.0, 0.1, 0.01, 0.001, 0.0001]) # Step size

# Store absolute value of difference of logq. Rows are different deltas, columns are different alphas.
pot_diff_thug = np.zeros((len(deltas), len(alphas)))

for ix, delta in enumerate(deltas):
    for jx, alpha in enumerate(alphas):
        avg = 0.0
        for i in range(M):
            x = THUG(x0, v0s, delta, alpha, q, logpi, norm_grad_log_pi)
            avg += np.abs(logpi(x) - logpi(x0)) / M
        pot_diff_thug[ix, jx] = avg
```

```python
# FIT LINEAR REGRESSION TO ALPHA=0.9 
model_thug = LinearRegression().fit(np.log10(deltas).reshape(-1, 1), np.log10(pot_diff_thug[:, -1]))
```

```python
fig, ax = plt.subplots()
lines = ax.plot(deltas, pot_diff_thug, marker='.', lw=2)
ax.set_xlabel(r'Log Scale $\delta$', fontsize=20)
ax.set_ylabel("Log Scale " + r'$|\ell(x_1) - \ell(x_0)|$', fontsize=12)
ax.set_xscale('log')
ax.set_yscale('log')
line = ax.plot(logspace, (10**model_thug.intercept_)*(logspace**model_thug.coef_), color='purple', label='Fitted Line - Slope {:.3}'.format(model_thug.coef_[0]))
#ax.plot(logspace, (10**model.intercept_)*(logspace**model.coef_), color='orange', label='Fitted Line - Slope {:.3}'.format(model.coef_[0]))
ax.set_title("Change in Potential Energy - THUG", fontsize=20)
plt.legend(lines + line, alphas.tolist() + ['Fitted Line - Slope {:.3}'.format(model_thug.coef_[0])])
plt.show()
```
