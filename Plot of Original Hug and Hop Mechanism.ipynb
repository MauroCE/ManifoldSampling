```python
# Run Hug Dynamics
def HugDynamic(x, T, B, v, grad_log_pi):
    δ = T / B
    positions = x
    for _ in range(B):
        # Move
        x = x + δ*v/2 
        positions = vstack((positions, x))
        # Reflect
        g = grad_log_pi(x)
        ghat = g / norm(g)
        v = v - 2*(v @ ghat) * ghat
        # Move
        x = x + δ*v/2
        positions = vstack((positions, x))
    return positions
```
