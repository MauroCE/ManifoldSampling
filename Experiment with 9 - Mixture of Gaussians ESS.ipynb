```python
p=6
d=2
mus = np.random.uniform(low=-1.0, high=1.0, size=(p, d))
random_matrices = uniform(low=-2.0, high=2.0, size=(p, d, d))
Sigmas = np.stack([random_matrices[i] @ random_matrices[i].T for i in range(p)])
coefs = np.repeat(1/p, p)

# Gaussian Components
distributions = [multivariate_normal(mu, Sigma) for (mu, Sigma) in zip(mus, Sigmas)]

# Mixture, log mixture, grad of log mixture
MG = lambda xy: np.sum(np.vstack([c * MVN.pdf(xy)  for (c, MVN) in zip(coefs, distributions)]), axis=0)
logpi_MG = lambda xy: np.log(MG(xy))
grad_log_pi_MG = lambda xy: (1 / MG(xy)) * np.sum(np.vstack([- c * multivariate_normal(mu, Sigma).pdf(xy) * solve(Sigma, xy - mu) for (c, mu, Sigma) in zip(coefs, mus, Sigmas)]), axis=0)

def sample_MG():
    """Samples from Mixture of Gaussians."""
    return distributions[np.where(rand() <= np.cumsum(np.repeat(1/p, p)))[0][0]].rvs()

q = multivariate_normal(np.zeros(d), np.eye(d))

# Find an odd contour
plt.contour(*prep_contour([-4, 4], [-4, 4], 0.01, MG))
plt.show()
```

```python
T = 1.0
B = 5
n = 1
N = 10000            # Notice this is the number of HMC step and half the number of HH steps
M = np.eye(d)
K = 10000            # Number of iterations used for preconditioning
lam = 2.0 #0.3 #1.5
k = 0.25 #0.25
alpha=0.5
```

```python
x0 = sample_MG()
```

```python
# Hug and Hop
t = time.time()
hhsamples, a11, a12 = run_hug_hop(x0, T, B, N, lam, k, q, logpi_MG, grad_log_pi_MG)
t_hh = time.time() - t
```

```python
# Thug and Hop
t = time.time()
thsamples, a41, a42 = run_thug_hop(x0, T, B, N, alpha, lam, k, q, logpi_MG, grad_log_pi_MG)
t_th = time.time() - t
```

```python
print("HH - Hug AR: {:.4}%".format(np.mean(a11) * 100))
print("HH - Hop AR: {:.4}%".format(np.mean(a12) * 100))
print("-"*20)
print("TH - Thug AR: {:.4}%".format(np.mean(a41) * 100))
print("TH - Hop AR:  {:.4}%".format(np.mean(a42) * 100))
```
