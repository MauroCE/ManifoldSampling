```python
import numpy as np
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt
from utils import logp as logp_scale
from Zappa.zappa import zappa_sampling_multivariate, zappa_sampling
from Zappa.ZappaStore import Zappa
from Manifolds.RotatedEllipse import RotatedEllipse
from utils import quick_MVN_scatter, MVN_scatters, prep_contour
from numpy.random import rand, uniform
from numpy.linalg import norm, solve, inv, det
from scipy.optimize import fsolve, root
from Manifolds.Manifold import Manifold
from tangential_hug_functions import Hug, HugTangential, NoAR
from utils import quick_3d_scatter, quick_3d_scatters, MVN_scatters, prep_contour
import plotly.graph_objects as go
```

```python
mu = np.zeros(2)
Sigma = np.array([[1.0, 0.99], [0.99, 2.0]])
target = multivariate_normal(mu, Sigma)
q = multivariate_normal(np.zeros(2), np.eye(2))
x0 = target.rvs()
z0 = target.pdf(x0)
```

```python
eps = 0.000001
T = 1.0
B = 5
N = 5000
alpha = 0.95
```

# Uniform Density

```python
# The target density now is decoupled from the gradient. The log density will be uniform on an epsilon ball 
def logpi(xy):
    with np.errstate(divide='ignore'):
        return np.log(abs(target.pdf(xy) - z0) <= eps)
    
# Gradient function is the gradient of the manifold!! In this case, since the manifold is the
# contour of a MVN, it will still be the usual gradient.
grad_func = lambda xy: - solve(Sigma, xy)

### Sampling
hug_samples, ahug = Hug(x0, T, B, N, q, logpi, grad_func)
thug_samples, athug = HugTangential(x0, T, B, N, alpha, q, logpi, grad_func)

### plotting
xlims = [min(np.min(hug_samples[:, 0]), np.min(thug_samples[:, 0]))*1.2, 
         max(np.max(hug_samples[:, 0]), np.max(thug_samples[:, 0]))*1.2]
ylims = [min(np.min(hug_samples[:, 1]), np.min(thug_samples[:, 1]))*1.2, 
         max(np.max(hug_samples[:, 1]), np.max(thug_samples[:, 1]))*1.2]

# Grid of points of contour lines
step = 0.01
x, y = np.mgrid[xlims[0]:xlims[1]:step, ylims[0]:ylims[1]:step]
pos = np.dstack((x, y))

labels = ['Hug: {:.3}%'.format(np.sum(ahug) * 100 / N), 'THug: {:.3}%'.format(np.sum(athug) * 100 / N)]

_, ax = MVN_scatters([hug_samples, thug_samples], target, xlims, ylims, levels=[z0], lw=2, labels=labels, return_axes=True, colors='purple')
#ax.contour(x, y, target.pdf(pos), linewidths=2, linestyles='dashed', levels=[z0+eps], alpha=0.5, zorder=1, colors='gray')
ax.contour(x, y, target.pdf(pos), linewidths=2, linestyles='dashed', levels=[z0-eps], alpha=0.5, zorder=1, colors='gray')
plt.title("Uniform Density (Hard Boundary $\\epsilon$ = {}) - Hug vs THUG ($\\alpha$ = {})".format(eps, alpha))
plt.show()
```

# Gaussian Kernel

We consider a Gaussian kernel $p:\mathbb{R}^n\to\mathbb{R}_{+}$ of the form
$$
p(x) = \mathcal{N}(\pi(x)\mid 0, \sigma^2) = \frac{1}{\sigma\sqrt{2\pi}}\exp\left\{-\frac{\pi(x)^2}{2\sigma^2}\right\}
$$

Hence with potential

$$
\log p(x) \propto - \frac{\pi(x)^2}{2\sigma^2}
$$

```python
alpha=0.99
sigma = 0.01

def logpi_gaussian(xy):
    return target.pdf(xy)**2 / (2 * sigma**2)

hug_samples, ahug = Hug(x0, T, B, N, q, logpi_gaussian, grad_func)
thug_samples, athug = HugTangential(x0, T, B, N, alpha, q, logpi_gaussian, grad_func)

xlims = [min(np.min(hug_samples[:, 0]), np.min(thug_samples[:, 0]))*1.2, 
         max(np.max(hug_samples[:, 0]), np.max(thug_samples[:, 0]))*1.2]
ylims = [min(np.min(hug_samples[:, 1]), np.min(thug_samples[:, 1]))*1.2, 
         max(np.max(hug_samples[:, 1]), np.max(thug_samples[:, 1]))*1.2]

step = 0.01
# Grid of points for contour plot
x, y = np.mgrid[xlims[0]:xlims[1]:step, ylims[0]:ylims[1]:step]
pos = np.dstack((x, y))

labels = ['Hug: {:.3}%'.format(np.sum(ahug) * 100 / N), 'THug: {:.3}%'.format(np.sum(athug) * 100 / N)]

_, ax = MVN_scatters([hug_samples, thug_samples], target, xlims, ylims, levels=[z0], lw=5, labels=labels, return_axes=True, colors='purple')
ax.contour(x, y, target.pdf(pos), linewidths=2, linestyles='dashed', levels=[z0+eps], alpha=0.5, zorder=1, colors='gray')
ax.contour(x, y, target.pdf(pos), linewidths=2, linestyles='dashed', levels=[z0-eps], alpha=0.5, zorder=1, colors='gray')
plt.title("Gaussian Kernel (soft boundary $\\sigma$={}) - Hug vs THUG (alpha={})".format(sigma, alpha))
plt.show()
```

```python
### Parameters of Mixture
d = 6
# Means
mus = np.random.uniform(low=-2.0, high=2.0, size=(d, 2))
random_matrices = uniform(low=-3.0, high=3.0, size=(d, 2, 2))
Sigmas = np.stack([random_matrices[i] @ random_matrices[i].T for i in range(d)])

# Mixture coefficients are all the same
coefs = np.repeat(1/d, d)

# Gaussian Components
distributions = [multivariate_normal(mu, Sigma) for (mu, Sigma) in zip(mus, Sigmas)]

# Mixture, log mixture, grad of log mixture
MG = lambda xy: np.sum(np.vstack([c * MVN.pdf(xy)  for (c, MVN) in zip(coefs, distributions)]), axis=0)
logpi = lambda xy: np.log(MG(xy))
grad_MG = lambda xy: (1 / MG(xy)) * np.sum(np.vstack([- c * multivariate_normal(mu, Sigma).pdf(xy) * solve(Sigma, xy - mu) for (c, mu, Sigma) in zip(coefs, mus, Sigmas)]), axis=0)

def sample_MG():
    """Samples from Mixture of Gaussians."""
    return distributions[np.where(rand() <= np.cumsum(np.repeat(1/d, d)))[0][0]].rvs()

# Find an odd contour
z = 0.016   # Sharp Turn: 0.016, Softer Turn: 0.004
plt.contour(*prep_contour([-4, 4], [-4, 4], 0.01, MG), levels=[z])
plt.show()
```

```python
x0_MG = sample_MG()
z0_MG = MG(x0_MG)
```
