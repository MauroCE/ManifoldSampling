```python
import time
import numpy as np
from numpy import zeros, diag, eye
from numpy.linalg import solve
from numpy.random import uniform
from scipy.stats import multivariate_normal
from tangential_hug_functions import Hug, Hop, HugTangential
from utils import quick_MVN_scatter, ESS, quick_MVN_marginals_kdes
import matplotlib.pyplot as plt
```

```python
d = 2
mu = zeros(d)
Sigma = diag(uniform(1.0, 10.0, size=d))  
target = multivariate_normal(mu, Sigma)
logpi = target.logpdf
grad_log_pi = lambda xy: - solve(Sigma, xy - mu)
q = multivariate_normal(zeros(d), eye(d))
x0 = target.rvs()
```

```python
T = 5.0              # Total integration time for HUG/THUG
B = 5                # Number of bounces per HUG/THUG step
N = 100              # Number of HUG samples per step
M = 10000             # Number of kernel iterations
lam = 2.0            # \lambda parameter for HOP
k = 0.25             # \kappa parameter for HOP
alpha=0.9            # Proportion of gradient to remove in THUG
```

# HUG + HOP

```python
def run_hug_hop(N):
    """Runs M iterations (cycles) of Hug and Hop. N is the number of Hug sub-iterations
    per iteration. In Hug&Hop paper N = 1. """
    samples = x = x0
    for _ in range(M):
        hug_samples, _ = Hug(x, T, B, N, q, logpi, grad_log_pi)    # N iterations of Hug
        x, _  = Hop(hug_samples[-1], lam, k, logpi, grad_log_pi)   # 1 iteration of Hop
        samples = np.vstack((samples, hug_samples, x))             # Store N + 1 samples
    return samples[1:]
```

```python
hh1   = run_hug_hop(1)
hh10  = run_hug_hop(10)
hh100 = run_hug_hop(100)
```

# THUG + HOP

```python
def run_thug_hop(N):
    """Runs M iterations (cycles) of THug and Hop. N is the number of Hug sub-iterations
    per iteration. In Hug&Hop paper N = 1. """
    samples = x = x0
    for _ in range(M):
        thug_samples, _ = HugTangential(x, T, B, N, alpha, q, logpi, grad_log_pi)    # N iterations of THug
        x, _  = Hop(thug_samples[-1], lam, k, logpi, grad_log_pi)   # 1 iteration of Hop
        samples = np.vstack((samples, thug_samples, x))             # Store N + 1 samples
    return samples[1:]
```

```python
def run_thug_hop(N, K):
    """Runs M iterations (cycles) of THug and Hop. N is the number of Hug sub-iterations
    per iteration. In Hug&Hop paper N = 1. """
    samples = x = x0
    while samples.shape[0] <= K:
        thug_samples, _ = HugTangential(x, T, B, N, alpha, q, logpi, grad_log_pi)    # N iterations of THug
        x, _  = Hop(thug_samples[-1], lam, k, logpi, grad_log_pi)   # 1 iteration of Hop
        samples = np.vstack((samples, thug_samples, x))             # Store N + 1 samples
    return samples[1:K+1]
```
