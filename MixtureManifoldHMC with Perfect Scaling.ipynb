```python
import numpy as np
from scipy.stats import multivariate_normal, norm
from utils import logf, logp, normalize, logf_Jacobian
from Manifolds.RotatedEllipse import RotatedEllipse
from HMC.gaussian_hmc import GaussianTargetHMC
from zappa import zappa_sampling, project
import matplotlib.pyplot as plt
from math import degrees, radians
from numpy.linalg import svd
from math import sqrt, log, exp
```

### Target Distribution

```python
# Parameters
mu = np.zeros(2)                                  # Center of the MVN
Sigma = np.array([[1.0, 0.99], [0.99, 2.0]])        # Covariance matrix of MVN
target = multivariate_normal(mean=mu, cov=Sigma)  # 2D distribution whose contours will be our manifolds

# Maximum level set value
z_max = target.pdf(mu)
```

### Settings

```python
N = 2000000
n = 1
m = 300
alpha = 0.2
T = 5 
epsilon = 0.1    
tol = 1.48e-08
a_guess = 1.0
s = 0.5
M = np.eye(2)
```

# Constant Scaling

### Uniform Distribution, Constant Scaling

```python
logp_new = lambda xy: logp(xy, sigma=s)
x = target.rvs()
z = target.pdf(x)
uc_samples = x
while len(uc_samples) < N: 

    # With probability alpha do n steps of HMC 
    if np.random.rand() <= alpha:
        new_samples = GaussianTargetHMC(q0=x, n=n, M=M, T=T, epsilon=epsilon, Sigma=Sigma, mu=mu).sample()

    # With probability 1 - alpha do m steps of Zappa's algorithm
    else:
        new_samples = zappa_sampling(x, RotatedEllipse(mu, Sigma, z), logf, logp_new, m, s, tol, a_guess)

    uc_samples = np.vstack((uc_samples, new_samples))
    x = new_samples[-1]
    z = target.pdf(x)
```

```python
xlow = -2.0
xhigh = 6.0
ylow = -3.0
yhigh = 5.0
x, y = np.mgrid[xlow:xhigh:.01, ylow:yhigh:.01]
pos = np.dstack((x, y))

fig, ax = plt.subplots(figsize=(20, 8))
ax.contour(x, y, target.pdf(pos), linewidths=5, alpha=0.3) 
ax.scatter(*uc_samples.T)
plt.show()
```

```python
marginal_x = lambda x: norm(loc=mu[0], scale=np.sqrt(Sigma[0, 0])).pdf(x)
marginal_y = lambda y: norm(loc=mu[1], scale=np.sqrt(Sigma[1, 1])).pdf(y)

xx = np.linspace(-4, 4, num=100)

fig, ax = plt.subplots(ncols=2, figsize=(20, 5))
ax[0].plot(xx, marginal_x(xx))
_ = ax[0].hist(uc_samples[:, 0], density=True, bins=50)
ax[1].plot(xx, marginal_y(xx))
_ = ax[1].hist(uc_samples[:, 1], density=True, bins=50)
```

### Jacobian, Constant Scaling

```python
logf_new = lambda xy: logf_Jacobian(xy, Sigma)
logp_new = lambda xy: logp(xy, sigma=s)

x = target.rvs()
z = target.pdf(x)
jc_samples = x
while len(jc_samples) < N: 

    # With probability alpha do n steps of HMC 
    if np.random.rand() <= alpha:
        new_samples = GaussianTargetHMC(q0=x, n=n, M=M, T=T, epsilon=epsilon, Sigma=Sigma, mu=mu).sample()

    # With probability 1 - alpha do m steps of Zappa's algorithm
    else:
        new_samples = zappa_sampling(x, RotatedEllipse(mu, Sigma, z), logf_new, logp_new, m, s, tol, a_guess)

    jc_samples = np.vstack((jc_samples, new_samples))
    x = new_samples[-1]
    z = target.pdf(x)
```

# Adaptive Scaling

```python
# Is 0.5 good for 0.9 * z_max?
s = 0.5
logp_new = lambda xy: logp(xy, sigma=s)

# Grab maximum contour value
z_max = target.pdf(mu)

# See if s=0.5 works for 0.9 * z_max
z_ok = 0.9 * z_max

# Grab a point on that manifold
manifold = RotatedEllipse(mu, Sigma, z_ok)
x0 = manifold.to_cartesian(np.random.uniform(0, 2*np.pi))

# Zappa 
samples_ok = zappa_sampling(x0, manifold, logf, logp_new, 200, s, tol, a_guess)

# Plot
fig, ax = plt.subplots(figsize=(20, 8))
ax.scatter(*samples_ok.T)
plt.show()
```

### Uniform, Adaptive Scaling

```python
# Grab ok perimeter
peri_ok = RotatedEllipse(mu, Sigma, z_ok).peri()

# Starting point of the algorithm
x = target.rvs()
z = target.pdf(x)
ua_samples = x
while len(ua_samples) < N: 
    # With probability alpha do n steps of HMC 
    if np.random.rand() <= alpha:
        new_samples = GaussianTargetHMC(q0=x, n=n, M=M, T=T, epsilon=epsilon, Sigma=Sigma, mu=mu).sample()

    # With probability 1 - alpha do m steps of Zappa's algorithm
    else:
        # Grab manifold
        manifold = RotatedEllipse(mu, Sigma, z)
        # Compute adaptive scale
        scale = (0.5 * manifold.peri() / peri_ok)
        # New function with adaptive scale
        logp_new = lambda xy: logp(xy, sigma=scale)
        new_samples = zappa_sampling(x, manifold, logf, logp_new, m, scale, tol, a_guess)

    ua_samples = np.vstack((ua_samples, new_samples))
    x = new_samples[-1]
    z = target.pdf(x)
```

```python
xlow = -2.0
xhigh = 6.0
ylow = -3.0
yhigh = 5.0
x, y = np.mgrid[xlow:xhigh:.01, ylow:yhigh:.01]
pos = np.dstack((x, y))

fig, ax = plt.subplots(figsize=(20, 8))
ax.contour(x, y, target.pdf(pos), linewidths=5, alpha=0.3) 
ax.scatter(*ua_samples.T)
plt.show()
```

```python
marginal_x = lambda x: norm(loc=mu[0], scale=np.sqrt(Sigma[0, 0])).pdf(x)
marginal_y = lambda y: norm(loc=mu[1], scale=np.sqrt(Sigma[1, 1])).pdf(y)

xx = np.linspace(-4, 4, num=100)

fig, ax = plt.subplots(ncols=2, figsize=(20, 5))
ax[0].plot(xx, marginal_x(xx))
_ = ax[0].hist(ua_samples[:, 0], density=True, bins=50)
ax[1].plot(xx, marginal_y(xx))
_ = ax[1].hist(ua_samples[:, 1], density=True, bins=50)
```

### Jacobian, Adaptive Scaling

```python
# Grab ok perimeter
peri_ok = RotatedEllipse(mu, Sigma, z_ok).peri()

# Starting point of the algorithm
x = target.rvs()
z = target.pdf(x)
ja_samples = x
while len(ja_samples) < N: 
    # With probability alpha do n steps of HMC 
    if np.random.rand() <= alpha:
        new_samples = GaussianTargetHMC(q0=x, n=n, M=M, T=T, epsilon=epsilon, Sigma=Sigma, mu=mu).sample()

    # With probability 1 - alpha do m steps of Zappa's algorithm
    else:
        # Grab manifold
        manifold = RotatedEllipse(mu, Sigma, z)
        # Compute adaptive scale
        scale = (0.5 * manifold.peri() / peri_ok)
        # New function with adaptive scale
        logp_new = lambda xy: logp(xy, sigma=scale)
        new_samples = zappa_sampling(x, manifold, logf_new, logp_new, m, scale, tol, a_guess)

    ja_samples = np.vstack((ja_samples, new_samples))
    x = new_samples[-1]
    z = target.pdf(x)
```

```python
xlow = -2.0
xhigh = 6.0
ylow = -3.0
yhigh = 5.0
x, y = np.mgrid[xlow:xhigh:.01, ylow:yhigh:.01]
pos = np.dstack((x, y))

fig, ax = plt.subplots(figsize=(20, 8))
ax.contour(x, y, target.pdf(pos), linewidths=5, alpha=0.3) 
ax.scatter(*ja_samples.T)
plt.show()
```
