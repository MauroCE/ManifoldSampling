```python
import numpy as np
from numpy import zeros, eye, array, linspace, exp, logspace, log, log10, vstack
from numpy.random import randn, default_rng
from numpy.linalg import solve, svd, norm

from scipy.linalg import qr
from scipy.stats import multivariate_normal as MVN
from scipy.optimize import fsolve

import matplotlib.pyplot as plt
from matplotlib import rc

from utils import prep_contour, line_perp_v_through_point, line_between
from Manifolds.GeneralizedEllipse import GeneralizedEllipse
from tangential_hug_functions import Hug
```

```python
rc('font',**{'family':'STIXGeneral'})
```

```python
μ = zeros(2)
Σ = array([[3, 0], [0, 1]])#eye(2)
π = MVN(μ, Σ)
f = π.logpdf
grad_log_π = lambda x: - solve(Σ, x - μ)
```

```python
z = -3
ellipse = GeneralizedEllipse(μ, Σ, exp(z))
ellipse_contour = prep_contour([-5, 5], [-5, 5], 0.01, f)
```

```python
def find_point_on_ellipse_with_given_x(xinit, yguess, z0=z):
    func   = lambda y: f(array([xinit, float(y)])) - z0
    yfound = fsolve(func, yguess)
    return array([xinit, *yfound])
```

```python
x0 = find_point_on_ellipse_with_given_x(-1, 2)
```

```python
v0 = array([0.7, 1.5])
```

```python
δ = 2
```

```python
x_mid = x0 + (δ/2)*v0
```

```python
z_mid = f(x_mid)
```

```python
g_mid = grad_log_π(x_mid)
```

```python
g_mid_hat = g_mid / norm(g_mid)
```

```python
comp_v0_along_g_mid_hat = g_mid_hat * (g_mid_hat @ v0)
```

```python
comp_v0_along_tangent = v0 - comp_v0_along_g_mid_hat
```

```python
v_new = v0 - 2*comp_v0_along_g_mid_hat
```

```python
x1 = x_mid + (δ/2)*v_new
```

```python
xvalues_hyperplane = linspace(-2, 1.5, 100)
yvalues_hyperplane = line_perp_v_through_point(g_mid_hat, x_mid, xvalues_hyperplane)
```

```python
SPOINTS = 100
LW_POINTS = 2

fig, ax = plt.subplots(figsize=(15, 4))
# Contours
ax.contour(*ellipse_contour, levels=[z], colors=['lightgray'], linestyles='-', zorder=0)
ax.contour(*ellipse_contour, levels=[z_mid], colors=['lightgray'], linestyles='-', zorder=0)
# Points
ax.scatter(*x0, s=SPOINTS, zorder=10, edgecolor='black', color='gray', lw=LW_POINTS)
ax.scatter(*x_mid, s=SPOINTS, zorder=10, edgecolor='navy', color='dodgerblue', lw=LW_POINTS)
ax.scatter(*x1, s=SPOINTS, zorder=10, edgecolor='brown', color='lightcoral', lw=LW_POINTS)
# Arrows
ax.arrow(*x0, *v0, length_includes_head=True, head_width=0.15)
ax.arrow(*x_mid, *g_mid_hat, length_includes_head=True, head_width=0.15)
ax.arrow(*x_mid, *v0, length_includes_head=True, head_width=0.15)
ax.arrow(*x_mid, *comp_v0_along_g_mid_hat, length_includes_head=True, head_width=0.15)
ax.arrow(*x_mid, *v_new, length_includes_head=True, head_width=0.15, overhang=0.7)
# Planes
ax.plot(xvalues_hyperplane, yvalues_hyperplane, color='lightgray')
ax.set_xlim(-4, 4)
ax.set_ylim(0.5, 4.5)
ax.set_aspect('equal')
plt.show()
```
