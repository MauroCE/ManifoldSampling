```python
import numpy as np
from numpy import zeros, eye
from numpy.linalg import norm

import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

from scipy.stats import multivariate_normal as MVN
from scipy.optimize import fsolve
```

```python
# Settings
r = 1.0     # Radius of sphere
h = 3.0     # Scale for SPS velocity
δ = 4.0     # Step-size for HUG

def point_on_sphere(r=r):
    point = MVN(zeros(3), eye(3)).rvs()
    return (point / norm(point)) * r

def sample_velocity():
    return MVN(zeros(3), eye(3)).rvs()

def is_on_sphere(z, r=r):
    return (norm(z) - r) == 0.0
```

```python
z = point_on_sphere()
v = sample_velocity()
```

```python
def sps_proposal(z, v, h=h):
    dz̃ = h*v
    dz = dz̃ - z*(z@dz̃) / (norm(z)**2)
    return (z + dz) * norm(z) / norm(z + dz)
```

```python
def hug_proposal(z, v, δ=δ):
    midpoint = z + (δ/2)*v
    ĝ = midpoint / norm(midpoint)
    v_new = v - 2 * ĝ * (ĝ @ v)
    return midpoint + (δ/2) * v_new
```

```python
def plot_δopt_for_h_grid():
    hs = np.linspace(0.0, 500.0, num=50)
    δs = []
    for h in hs:
        find_δ = lambda δ: norm(sps_proposal(z, v, h=h) - hug_proposal(z, v, δ))
        δopt = fsolve(find_δ, 1.0)
        δs.append(δopt)
    δs = np.array(δs)
    return hs, δs
```

```python
hs, δs = plot_δopt_for_h_grid()
```

# Comparing Their and Mine Sphere sampling

```python
from numpy import array, zeros, eye, sqrt
from numpy.random import rand
from scipy.stats import multivariate_normal as MVN
```

```python
d  = 2                      # Dimension of the Sphere. Ambient space is R^{d+1}
R  = 1                      # Parameter
```

```python
def generate_north_pole(d):
    """Generates d-dimensional North Pole."""
    north_pole     = zeros(d + 1)
    north_pole[-1] = 1
    return north_pole

def sp(z, R=R):
    """Takes a point z on the d-dimensional sphere with ambient space R^{d+1}, and maps it to R^d."""
    return R * (z[:-1] / (1 - z[-1]))

def Jsp(x, R=R):
    """Jacobian of Stereographic Projection."""
    return (R**2 + x@x)**(len(x))

def sp_inv(x, R=R):
    """Inverse Stereographic Projection."""
    return r_[2*R*x / (x@x + R**2), (x@x - R**2) / (x@x + R**2)]

def πsp(x, R=R):
    """Transformed density. Density on Sphere."""
    return π(x) * Jsp(x, R=R)

def SPS(x, R, h, πsp):
    """Stereographic Projection Sampler."""
    d  = len(x)
    z  = sp_inv(x, R=R)
    dz̃ = MVN(zeros(d+1), (h**2)*eye(d+1))
    dz - (z @ dz̃)*z/(z@z)
    ẑ  = (z + dz)
    ẑ /= sqrt(ẑ @ ẑ)
    x̂  = sp(ẑ, R=R)
    a  = min(1, (π(x̂)*(R**2 + x̂@x̂)**d) / (π(x)*(R**2 + x@x)**d))
    u  = rand()
    if u <= a:
        return x̂
    else:
        return x
```
