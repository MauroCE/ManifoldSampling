```python
import numpy as np
from scipy.stats import multivariate_normal
from Zappa.zappa import zappa_projectv, zappa_sampling
from Manifolds.RotatedEllipse import RotatedEllipse
from utils import logf_Jacobian
from utils import logp as logp_scale
from utils import quick_MVN_scatter, quick_MVN_marginals
import matplotlib.pyplot as plt
from utils import normalize
from Zappa.zappa import project
```

```python
mu = np.zeros(2)
Sigma = np.array([[1.0, 0.99], [0.99, 2.0]])
target = multivariate_normal(mu, Sigma)
s = 0.5
n = 50
tol = 1.48e-08
x = target.rvs()
z = target.pdf(x)
ellipse = RotatedEllipse(mu, Sigma, z)
logf = lambda xy: logf_Jacobian(xy, Sigma)
logp = lambda xy: logp_scale(xy, s)
```

```python
samples_normal = zappa_sampling(x, ellipse, logf, logp, n, s, tol, 1.0)
```

```python
quick_MVN_scatter(samples_normal, target, levels=[z], alpha=0.5)
```

```python
samples = zappa_projectv(x, ellipse, logf, logp, n, s, tol, 1.0)
```

```python
quick_MVN_scatter(samples, target, levels=[z], alpha=0.5)
```

```python
plt.scatter(*samples.T, c=np.arange(len(samples)))
```

```python
maxiter=50
a_guess=1.0
logu = np.log(np.random.rand())
s=0.2
x = target.rvs()
z = target.pdf(x)
ellipse = RotatedEllipse(mu, Sigma, z)
```

```python
manifold = ellipse
Qx = manifold.Q(x)                       # Gradient at x.                             Size: (d + m, )
gx_basis = normalize(Qx)                 # ON basis for gradient at x                 Size: (d + m, )
tx_basis = manifold.tangent_basis(Qx)    # ON basis for tangent space at x using SVD  Size: (d + m, d)
v_sample = s*np.random.randn(1)  # Isotropic MVN with scaling sigma         Size: (d, )
v = tx_basis @ v_sample    # Linear combination of the basis vectors  Size: (d + m, )
a, flag = project(x, v, Qx, manifold.q, tol, a_guess, maxiter)
if flag == 0:                                  # Projection failed                        
    # Sample new velocity at x.
    v_sample = s*np.random.randn(1)  # Isotropic MVN with scaling sigma         Size: (d, )
    v = tx_basis @ v_sample    # Linear combination of the basis vectors  Size: (d + m, )
    print("Failed")
y = x + v + a*Qx.flatten()                              # Projected point (d + m, )
Qy = manifold.Q(y)                         # Gradient at y.                            Size: (d + m, )
gy_basis = normalize(Qy)             # ON basis for gradient at y                Size: (d + m, )
ty_basis = manifold.tangent_basis(Qy)      # ON basis for tangent space at y using SVD Size: (d + m, d)
v_prime_sample = (x - y) @ ty_basis  # Components along tangent                  Size: (d + m, )
if logu > logf(y) + logp(v_prime_sample) - logf(x) - logp(v_sample):
    print("mh failed")
# Backward Projection
v_prime = v_prime_sample @ ty_basis.T   # Linear combination of the basis vectors. Size: (d + m, )
a_prime, flag = project(y, v_prime, Qy, manifold.q, tol, a_guess, maxiter)
if flag == 0:
    print("reproj failed")
# Updated stuff
v_sample_new = (v) @ ty_basis
v_new = v_sample_new @ ty_basis.T
```

```python
xlims=[-2, 2]
ylims=[-1, 2]
step=0.01
xx, yy = np.mgrid[xlims[0]:xlims[1]:step, ylims[0]:ylims[1]:step]
pos = np.dstack((xx, yy))
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
ax.contour(xx, yy, target.pdf(pos), linewidths=5, levels=[z], alpha=0.5, zorder=1, colors='gray') 
ax.scatter(*x, label="x")
ax.arrow(*x, *gx_basis.flatten(), length_includes_head=True, width=0.00001)
ax.arrow(*x, *tx_basis.flatten(), length_includes_head=True, width=0.00001)
ax.scatter(*(x+v), label="x+v")
ax.scatter(*y, label="y")
ax.arrow(*y, *gy_basis.flatten(), length_includes_head=True, width=0.00001)
ax.arrow(*y, *ty_basis.flatten(), length_includes_head=True, width=0.00001)
ax.scatter(*(y+v_prime), label="v'")
ax.scatter(*(y+v), label="y+v")
ax.scatter(*(y+v_new), label="y + v_new")
ax.scatter(*boh, label="boh")
ax.scatter(*(y + ((v @ ty_basis) @ ty_basis.T)), label="boh2")
ax.legend()
#ax.set_xlim(-0.6, -0.4)
#ax.set_ylim(-2.45, -2.38)
```
