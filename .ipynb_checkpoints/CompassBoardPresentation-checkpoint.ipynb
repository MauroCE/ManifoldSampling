```python
import numpy as np
from numpy import zeros, eye, ones
from scipy.stats import multivariate_normal as MVN
import matplotlib.pyplot as plt
from utils import prep_contour
from matplotlib import rcParams
import matplotlib.colors as mcolors
```

```python
prior = MVN(zeros(2), eye(2))
logprior = lambda θ: prior.logpdf(θ)
sampleprior = lambda n=1: prior.rvs(n)
h = lambda θ: θ[0] + θ[1]
σ = 0.3
y = 1
μ = y * ones(2) / (2 + σ**2)
Σ = eye(2) - (1 / (2 + σ**2))*ones((2, 2))
π = MVN(μ, Σ)
logπ = π.logpdf
likelihood = lambda θ: MVN(h(θ), (σ**2)*eye(1))
likelihood_function = lambda θ: likelihood(θ).pdf(y)
log_likelihood = lambda θ: likelihood(θ).logpdf(y)
```

```python
posterior_samples = π.rvs(100000)
```

```python
θ0s = np.linspace(-3, 4, 100)
θ1s = -θ0s + y
```

```python
dens_cmap = mcolors.LinearSegmentedColormap.from_list(
    'whiteblue', [np.array((1.000, 1.000, 1.000, 1.)),
                  np.array((0.419, 0.682, 0.840, 1.)),
                  np.array((0.031, 0.188, 0.420, 1.))])
```

```python
θ_grid = np.stack(
    np.meshgrid(np.linspace(-4, 4, 100), np.linspace(-4, 4, 100))
)
```

## Manifold Only

```python
fig, ax = plt.subplots(figsize=(8, 8))
# Modify the axes
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('dimgrey')
ax.spines['bottom'].set_color('dimgrey')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
ax.set_xticks([-4, -3, -2, -1, 1, 2, 3, 4])
ax.set_yticks([-4, -3, -2, -1, 1, 2, 3, 4])
ax.tick_params(axis='x', color='dimgrey')
ax.tick_params(axis='y', color='dimgrey')
ax.xaxis.label.set_color('dimgrey')
ax.yaxis.label.set_color('dimgrey')
# Plot manifold
# ax.contour(*prep_contour((-4, 4), (-4, 4), 0.01, logprior), colors='lightgrey', linestyles='solid')
ax.plot(θ0s, θ1s, c='crimson', zorder=10, linewidth=3)
#ax.scatter(*posterior_samples.T, s=30, alpha=0.4, color='powderblue', edgecolor='cornflowerblue', zorder=5)
#ax.contour(*prep_contour((-4, 4), (-4, 4), 0.01, π.pdf), levels=[0.005, 0.01, 0.05, 0.1, 0.2, 0.3, 0.5])
#ax.pcolormesh(θ_grid[0], θ_grid[1], np.apply_along_axis(π.pdf, 0, θ_grid), cmap=dens_cmap)
# ax.pcolormesh(θ_grid[0], θ_grid[1], np.apply_along_axis(likelihood_function, 0, θ_grid), cmap=dens_cmap)
ax.set_aspect('equal')
ax.annotate(r'$y = \theta_1 + \theta_2$', (-2, 3.5), fontsize=15)
ax.grid(color='whitesmoke')
plt.savefig("figures/plot1_compasspresentation.png")
plt.show()
```

# Manifold and Likelihood

```python
fig, ax = plt.subplots(figsize=(8, 8))
# Modify the axes
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('dimgrey')
ax.spines['bottom'].set_color('dimgrey')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
ax.set_xticks([-4, -3, -2, -1, 1, 2, 3, 4])
ax.set_yticks([-4, -3, -2, -1, 1, 2, 3, 4])
ax.tick_params(axis='x', color='dimgrey')
ax.tick_params(axis='y', color='dimgrey')
ax.xaxis.label.set_color('dimgrey')
ax.yaxis.label.set_color('dimgrey')
# Plot manifold
ax.plot(θ0s, θ1s, c='crimson', zorder=10, linewidth=3)
ax.pcolormesh(θ_grid[0], θ_grid[1], np.apply_along_axis(likelihood_function, 0, θ_grid), cmap=dens_cmap)
ax.set_aspect('equal')
ax.annotate(r'$y \sim  \mathcal{N}(\theta_1 + \theta_2, \sigma^2)$', (-1.8, 3.5), fontsize=15)
ax.grid(color='whitesmoke', zorder=-1000)
plt.savefig("figures/plot2_compasspresentation.png")
plt.show()
```

## Manifold, Prior and Posterior

```python
fig, ax = plt.subplots(figsize=(8, 8))
# Modify the axes
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['left'].set_color('dimgrey')
ax.spines['bottom'].set_color('dimgrey')
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)
ax.set_xticks([-4, -3, -2, -1, 1, 2, 3, 4])
ax.set_yticks([-4, -3, -2, -1, 1, 2, 3, 4])
ax.tick_params(axis='x', color='dimgrey')
ax.tick_params(axis='y', color='dimgrey')
ax.xaxis.label.set_color('dimgrey')
ax.yaxis.label.set_color('dimgrey')
# Plot manifold
ax.contour(*prep_contour((-4, 4), (-4, 4), 0.01, logprior), colors='lightgrey', linestyles='solid')
ax.plot(θ0s, θ1s, c='crimson', zorder=10, linewidth=3)
#ax.scatter(*posterior_samples.T, s=30, alpha=0.4, color='powderblue', edgecolor='cornflowerblue', zorder=5)
#ax.contour(*prep_contour((-4, 4), (-4, 4), 0.01, π.pdf), levels=[0.005, 0.01, 0.05, 0.1, 0.2, 0.3, 0.5])
ax.pcolormesh(θ_grid[0], θ_grid[1], np.apply_along_axis(π.pdf, 0, θ_grid), cmap=dens_cmap)
ax.set_aspect('equal')
ax.annotate(r'$p(\theta\mid y)\propto p(\theta) p(y\mid\theta)\,\,\,\,\,\,\,\,\,\,\,y\sim \mathcal{N}(\theta_1+\theta_2, \sigma^2)$', (-2, 3.5), fontsize=15)
ax.annotate(r'$\theta\sim \mathcal{N}(0, I)$', (1.3, 3), fontsize=15)
ax.grid(color='whitesmoke')
plt.savefig("figures/plot3_compasspresentation.png")
plt.show()
```
