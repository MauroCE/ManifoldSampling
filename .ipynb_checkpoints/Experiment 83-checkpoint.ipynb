```python
def generate_settings(
    d=10, 
    N=5000,
    B=20,
    δ=0.1,
    δmin=1e-3,
    δmax=100.0,
    εmin=1e-16,
    min_pm=1e-4,
    maxiter=500,
    verbose=True,
    εs_fixed=None,
    adaptiveε=True,
    adaptiveδ=True,
    z0_manual=None,
    pm_target=0.23,
    pm_switch=0.05,
    prior_seed=1234,
    low_memory=True,
    integrator='rwm',
    εprop_switch=0.01,
    ε0_manual=None,
    quantile_value=0.9,
    initialization='prior',
    switch_strategy='pm',
    prior='uniform',
    kernel='uniform',
    prior_scale=7,
    thinning=10,
    burnin=100):
    # Generate the manifold
    μ = zeros(d)
    Σ = diag(np.r_[0.1, ones(d-1)])  
    level_set_value = -23
    manifold = GeneralizedEllipse(μ, Σ, exp(level_set_value), prior=prior, kernel=kernel, prior_scale=prior_scale)
    # grab arguments
    arguments = locals()
    return arguments
```

```python
# delta = 1.0
SETTINGS_RWM_1           = generate_settings(integrator='rwm', adaptiveδ=False, δ=1.0)
SETTINGS_THUG_1          = generate_settings(integrator='thug', adaptiveδ=False, δ=1.0)
SETTINGS_RWM_THEN_THUG_1 = generate_settings(integrator='rwm_then_thug', adaptiveδ=False, δ=1.0)

# delta = 0.05
SETTINGS_RWM_005           = generate_settings(integrator='rwm', adaptiveδ=False, δ=0.05)
SETTINGS_THUG_005          = generate_settings(integrator='thug', adaptiveδ=False, δ=0.05)
SETTINGS_RWM_THEN_THUG_005 = generate_settings(integrator='rwm_then_thug', adaptiveδ=False, δ=0.05)

# delta = 0.001
SETTINGS_RWM_0001           = generate_settings(integrator='rwm', adaptiveδ=False, δ=0.001)
SETTINGS_THUG_0001          = generate_settings(integrator='thug', adaptiveδ=False, δ=0.001)
SETTINGS_RWM_THEN_THUG_0001 = generate_settings(integrator='rwm_then_thug', adaptiveδ=False, δ=0.001)
```

# Plotting

```python
# Colors
SMC_RWM_COLOR = 'orange'     # warm, light
SMC_RT_COLOR = 'lightcoral'  # warm, light
MS_RWM_COLOR = 'thistle'   # cold, light
MS_RT_COLOR = 'lightskyblue' # cold, light
SMC_RT_LINE_COLOR = 'firebrick'  # warm, dark
MS_RT_LINE_COLOR = 'royalblue'   # cold, dark
```

```python
def prep_ess(SAMPLER, divide_by_n=True, ms=False):
    if divide_by_n:
        if not ms:
            return np.array(SAMPLER.ESS) / SAMPLER.N
        else:
            values = deepcopy(np.array(SAMPLER.ESS))
            values[0] = SAMPLER.N
            return values / SAMPLER.N
    else:
        return np.array(SAMPLER.ESS) / (SAMPLER.N*(SAMPLER.B+1))
```

```python
fig, ax = plt.subplots(ncols=3, nrows=3, figsize=(3*6, 3*3))
LIST_SMC_RWMS = [SMC_RWM_1, SMC_RWM_005, SMC_RWM_0001]
LIST_MS_RWMS  = [MS_RWM_1, MS_RWM_005, MS_RWM_0001]
LIST_SMC_RTS  = [SMC_RT_1, SMC_RT_005, SMC_RT_0001]
LIST_MS_RTS   = [MS_RT_1, MS_RT_005, MS_RT_0001]
MARKERSIZE=50
for i in range(3):
    # Tolerances
    ax[i, 0].plot(LIST_SMC_RWMS[i].εs, label='SMC-RWM', color=SMC_RWM_COLOR)
    ax[i, 0].plot(LIST_MS_RWMS[i].εs, label='MS-RWM', color=MS_RWM_COLOR)
    ax[i, 0].plot(LIST_SMC_RTS[i].εs, label='SMC-RT', color=SMC_RT_COLOR)
    ax[i, 0].plot(LIST_MS_RTS[i].ϵs, label='MS-RT', color=MS_RT_COLOR)
    ax[i, 0].axvline(x=LIST_SMC_RTS[i].n_switch-1, color=SMC_RT_LINE_COLOR, linestyle='--')
    ax[i, 0].axvline(x=LIST_MS_RTS[i].n_switch-1, color=MS_RT_LINE_COLOR, linestyle='--')
    ax[i, 0].scatter(LIST_SMC_RWMS[i].n, LIST_SMC_RWMS[i].εs[-1], marker='o', color=SMC_RWM_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 0].scatter(LIST_MS_RWMS[i].n, LIST_MS_RWMS[i].εs[-1], marker='o', color=MS_RWM_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 0].scatter(LIST_SMC_RTS[i].n, LIST_SMC_RTS[i].εs[-1], marker='o', color=SMC_RT_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 0].scatter(LIST_MS_RTS[i].n, LIST_MS_RTS[i].εs[-1], marker='o', color=MS_RT_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 0].set_yscale('log')
    ax[i, 0].set_title("ε")
    # Acceptance Probability
    ax[i, 1].plot(LIST_SMC_RWMS[i].ϵs, LIST_SMC_RWMS[i].APS, label='SMC-RWM', color=SMC_RWM_COLOR)
    ax[i, 1].plot(LIST_MS_RWMS[i].ϵs, LIST_MS_RWMS[i].PROP_MOVED, label='MS-RWM', color=MS_RWM_COLOR)
    ax[i, 1].plot(LIST_SMC_RTS[i].ϵs, LIST_SMC_RTS[i].APS, label='SMC-RT', color=SMC_RT_COLOR)
    ax[i, 1].plot(LIST_MS_RTS[i].ϵs, LIST_MS_RTS[i].PROP_MOVED, label='MS-RT', color=MS_RT_COLOR)
    ax[i, 1].axvline(x=LIST_SMC_RTS[i].εs[LIST_SMC_RTS[i].n_switch], color=SMC_RT_LINE_COLOR, linestyle='--')
    ax[i, 1].axvline(x=LIST_MS_RTS[i].εs[LIST_MS_RTS[i].n_switch], color=MS_RT_LINE_COLOR, linestyle='--')
    ax[i, 1].scatter(LIST_SMC_RWMS[i].ϵs[-1], LIST_SMC_RWMS[i].APS[-1], marker='o', color=SMC_RWM_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 1].scatter(LIST_MS_RWMS[i].ϵs[-1], LIST_MS_RWMS[i].PROP_MOVED[-1], marker='o', color=MS_RWM_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 1].scatter(LIST_SMC_RTS[i].ϵs[-1], LIST_SMC_RTS[i].APS[-1], marker='o', color=SMC_RT_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 1].scatter(LIST_MS_RTS[i].ϵs[-1], LIST_MS_RTS[i].PROP_MOVED[-1], marker='o', color=MS_RT_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 1].set_title("AP")
    ax[i, 1].set_xscale('log')
    # ESS
    ax[i, 2].plot(LIST_SMC_RWMS[i].ϵs, prep_ess(LIST_SMC_RWMS[i]), label='SMC-RWM', color=SMC_RWM_COLOR)
    ax[i, 2].plot(LIST_MS_RWMS[i].ϵs, prep_ess(LIST_MS_RWMS[i], ms=True), label='MS-RWM', color=MS_RWM_COLOR)
    ax[i, 2].plot(LIST_SMC_RTS[i].ϵs, prep_ess(LIST_SMC_RTS[i]), label='SMC-RT', color=SMC_RT_COLOR)
    ax[i, 2].plot(LIST_MS_RTS[i].ϵs, prep_ess(LIST_MS_RTS[i], ms=True), label='MS-RT', color=MS_RT_COLOR)
    ax[i, 2].axvline(x=LIST_SMC_RTS[i].εs[LIST_SMC_RTS[i].n_switch], color=SMC_RT_LINE_COLOR, linestyle='--')
    ax[i, 2].axvline(x=LIST_MS_RTS[i].εs[LIST_MS_RTS[i].n_switch], color=MS_RT_LINE_COLOR, linestyle='--')
    ax[i, 2].scatter(LIST_SMC_RWMS[i].ϵs[-1], prep_ess(LIST_SMC_RWMS[i])[-1], marker='o', color=SMC_RWM_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 2].scatter(LIST_MS_RWMS[i].ϵs[-1], prep_ess(LIST_MS_RWMS[i], ms=True)[-1], marker='o', color=MS_RWM_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 2].scatter(LIST_SMC_RTS[i].ϵs[-1], prep_ess(LIST_SMC_RTS[i])[-1], marker='o', color=SMC_RT_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 2].scatter(LIST_MS_RTS[i].ϵs[-1], prep_ess(LIST_MS_RTS[i], ms=True)[-1], marker='o', color=MS_RT_COLOR, s=MARKERSIZE, zorder=10)
    ax[i, 2].set_title("ESS/N")
    ax[i, 2].set_yscale('log')
    ax[i, 2].set_xscale('log')
    # Legend
    ax[i, 2].legend()
plt.show()
```
