### Imports

```python
import numpy as np
from scipy.linalg import norm, solve
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
from two_moons import tm_deterministic, tm_jacobian, tm_distance_gradient
from numpy.random import uniform, normal, rand
from scipy.stats import multivariate_normal
from pandas import DataFrame as df
from seaborn import pairplot
from scipy.stats import uniform as udist
from scipy.stats import norm as ndist
```

### Settings

```python
y_star = np.zeros(2)
T = 0.05       # Integration time for Hug
B = 5         # Number of steps
delta = T / B # Step size
N = 10000      # Number of Hug iterations
epsilon_spherical = 0.05
epsilon_tangential = 0.05
scale = 1.0     # Scale for the MVN in the spherical case. Equiv to changing B and delta
alpha = 0.6     # Proportion of the gradient to remove from spherical to make tangential
```

### Initial Point on Manifold

```python
# Sample randomly the seeds
a0 = uniform(low=-np.pi/2, high=np.pi/2)
r0 = normal(loc=0.1, scale=0.01)

# Start with a guess
guess = np.array([0, 0, a0, r0])

# Find point on manifold using optimization
func = lambda xi: np.r_[tm_deterministic(xi) - y_star, 0, 0]  # Append 0, 0 to make fsolve work.
xi0 = fsolve(func, guess)
```

### Spherical Hug

```python
def logprior(theta):
    """Computes log prior uniform for theta. Computes log p(\theta)"""
    if np.all(np.abs(theta) <= np.pi/2):
        return np.log(0.25)
    else:
        return -np.inf
    
def loglatent_prior(z):
    """Computes log p(z | theta). In our case log p(z)."""
    a, r = z
    return udist(loc=-np.pi/2, scale=np.pi).logpdf(a) + ndist(loc=0.1, scale=0.01).logpdf(r)

def log_joint(xi):
    """Computes log p(xi) = log p(theta) + log p(z)"""
    return logprior(xi[:2]) + loglatent_prior(xi[2:])
```

```python
def hug(xi0, B, delta, epsilon, scale):
    """Hug algorithm for sampling from distance manifold."""
    xi = xi0
    v_dist = multivariate_normal(mean=np.zeros(4), cov=scale * np.eye(4))
    v0 = v_dist.rvs()
    v = v0
    for b in range(B):
        # Move in direction of v
        xi = xi + (delta/2) * v
        # Compute normalized gradient at new position ####### COULD DO EARLY REJECTION HERE
        g = tm_distance_gradient(xi, y_star, norm(tm_deterministic(xi) - y_star)).flatten() 
        g = g / norm(g)
        # Reflect velocity
        v = v - 2 * (v @ g) * g
        # Move in direction of new velocity
        xi = xi + (delta / 2)*v
    # Check if proposed point is inside the bounded region
    if norm(tm_deterministic(xi) - y_star) > epsilon:
        return xi0, v0
    else:
        # Metropolis-Hastings acceptance
        if rand() <= np.exp(log_joint(xi) - log_joint(xi0) + v_dist.logpdf(v) - v_dist.logpdf(v0)):
            # Accept!
            return xi, v
        else:
            return xi0, v0
```

```python
samples = xi0
xi = xi0
for i in range(N):
    xi, _ = hug(xi, B, delta, epsilon_spherical, scale)
    samples = np.vstack((samples, xi))
```

### Tangential Hug

```python
def tangential_hug(xi0, B, delta, epsilon, scale, alpha):
    """Tangential Hug algorithm for sampling from distance manifold."""
    xi = xi0
    # Compute gradient here and normalize it
    g = tm_distance_gradient(xi, y_star, norm(tm_deterministic(xi) - y_star)).flatten()
    g = g / norm(g)
    # Rather than samplign spherically, want to sample tangentially
    v_dist_spherical = multivariate_normal(mean=np.zeros(4), cov=scale * np.eye(4))
    v0_spherical = v_dist_spherical.rvs()
    # Squeeze gradient direction
    v0 = v0_spherical - alpha * g * (g @ v0_spherical)  
    # Instantiate distribution for tilted MVN
    v0_dist = multivariate_normal(mean=np.zeros(4), cov=(np.eye(4) + alpha*(alpha-2)*np.outer(g, g)))
    v = v0
    for b in range(B):
        # Move in direction of v
        xi = xi + (delta/2) * v
        # Compute normalized gradient at new position ####### COULD DO EARLY REJECTION HERE
        g = tm_distance_gradient(xi, y_star, norm(tm_deterministic(xi) - y_star)).flatten() 
        g = g / norm(g)
        # Reflect velocity
        v = v - 2 * (v @ g) * g
        # Move in direction of new velocity
        xi = xi + (delta / 2)*v
    # Check if proposed point is inside the bounded region
    if norm(tm_deterministic(xi) - y_star) > epsilon:
        return xi0, v0
    else:
        # Metropolis-Hastings acceptance
        # Must use the velocity distribution at the new pair
        g = tm_distance_gradient(xi, y_star, norm(tm_deterministic(xi) - y_star)).flatten()
        g = g / norm(g)
        v_dist = multivariate_normal(mean=np.zeros(4), cov=(np.eye(4) + alpha*(alpha-2)*np.outer(g, g)))
        if rand() <= np.exp(log_joint(xi) - log_joint(xi0) + v_dist.logpdf(v) - v0_dist.logpdf(v0)):
            # Accept!
            return xi, v
        else:
            return xi0, v0
```

```python
t_samples = xi0
xi = xi0
for i in range(N):
    xi, _ = tangential_hug(xi, B, delta, epsilon_tangential, scale, alpha)
    t_samples = np.vstack((t_samples, xi))
```

```python
fig, ax = plt.subplots(figsize=(20,8), ncols=2)
ax[0].scatter(*samples[:, :2].T)
ax[0].set_aspect("equal")
ax[0].set_title("Spherical Hug (Unique: {:.3}%)".format((np.unique(samples, axis=0).shape[0] / N)*100))
ax[1].scatter(*t_samples[:, :2].T)
ax[1].set_aspect("equal")
ax[1].set_title("Tangential Hug (Unique: {:.3}%)".format((np.unique(t_samples, axis=0).shape[0] / N)*100))
plt.show()
```
