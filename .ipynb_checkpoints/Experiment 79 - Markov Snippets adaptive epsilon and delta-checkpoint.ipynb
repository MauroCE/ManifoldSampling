```python
def generate_setting(m, B, δ, N, ϵ0=20, thinning=10, thug=True, verbose=True, initialization='init_RWMϵ0', burnin=100, δ0=0.01, kernel_type='uniform', ϵmin=1e-4, quantile_value=0.01, maxiter=50, ϵprop_switch=0.05, pmoved_switch=0.15, switch_strategy='ap', δmin=1e-4, δmax=100.0, ap_target=0.3):
    """Generates an object from which one can grab the settings. This allows one to run multiple scenarios."""
    θ0        = array([3.0, 1.0, 2.0, 0.5])      # True parameter value on U(0, 10) scale.
    d         = 4 + m                            # Dimensionality of ξ=(θ, z)
    ystar     = data_generator(θ0, m, seed=1234) # Observed data
    q         = MVN(zeros(d), eye(d))            # Proposal distribution for THUG
    ξ0        = find_point_on_manifold_from_θ(ystar=ystar, θfixed=ndtri(θ0/10), ϵ=1e-5, maxiter=5000, tol=1e-15, kernel_type=kernel_type)
    resetwarnings()
    manifold  = GKManifold(ystar, kernel_type=kernel_type)
    if δ0 is None:
        δ0 = δ
    return {
        'θ0': θ0,
        'm' : m,
        'd' : d,
        'ystar': ystar,
        'q': q,
        'ξ0': ξ0,
        'B': B,
        'δ': δ,
        'N': N,
        'manifold': manifold,
        'thinning': thinning,
        'thug': thug,
        'verbose': verbose,
        'initialization': initialization,
        'burnin': burnin,
        'δ0': δ0,
        'kernel_type': kernel_type,
        'ϵmin': ϵmin,
        'quantile_value': quantile_value,
        'maxiter': maxiter,
        'ϵprop_switch': ϵprop_switch,
        'pmoved_switch': pmoved_switch,
        'switch_strategy': switch_strategy,
        'δmin': δmin,
        'δmax': δmax,
        'ap_target': ap_target,
        'ϵ0': ϵ0
    }
```

```python
N       = 5000
B       = 20
δ       = 0.1 #0.01
δ0      = 0.01
m       = 50
ϵmin    = 1e-4
ϵ0      = 20
maxiter = 200
thinning= 10
burnin  = 100
quantile_value = 0.01
kernel_type = 'uniform'
initialization = 'init_prior'
ϵprop_switch = 1e-4
switch_strategy = 'ap'
pmoved_switch = 0.15
δmin    = 1e-4
δmax    = 100.0
ap_target = 0.3
```

```python
SETTINGS50 = generate_setting(
    m=m, B=B, δ=δ, N=N, thinning=thinning, δ0=δ0, 
    kernel_type=kernel_type, thug=False, verbose=True, initialization=initialization,
    burnin=burnin, maxiter=maxiter, ϵmin=ϵmin, quantile_value=quantile_value, ϵprop_switch=ϵprop_switch,
    pmoved_switch=pmoved_switch, switch_strategy=switch_strategy, δmin=δmin, δmax=δmax, ap_target=ap_target)
```

### MS-RWM - adapt $\epsilon$ and $\delta$

```python
# Instantitate the algorithm
MSRWM_adaptiveδ = MSAdaptiveTolerancesAdaptiveδ(SETTINGS50)
# Sample
zP_RWM = MSRWM_adaptiveδ.sample()
```

### MS-THUG - adapt $\epsilon$ and $\delta$

The problem is that if we initialize from the prior then we are not in the filamentary regime, and so it does not make sense to use THUG. THUG will not be able to get any closer to the true manifold.

Instead, we try initialize from a relatively smallish epsilon and see what happens. 

```python
SETTINGS50_THUG = generate_setting(
    m=m, B=B, δ=δ, N=N, thinning=thinning, δ0=0.01, ϵ0=5,
    kernel_type=kernel_type, thug=True, verbose=True, initialization='init_THUGϵ0',
    burnin=burnin, maxiter=maxiter, ϵmin=1e-8, quantile_value=quantile_value, ϵprop_switch=ϵprop_switch,
    pmoved_switch=pmoved_switch, switch_strategy=switch_strategy, δmin=δmin, δmax=δmax, ap_target=ap_target)
```

```python
# Instantitate the algorithm
MSTHUG_adaptiveδ = MSAdaptiveTolerancesAdaptiveδ(SETTINGS50_THUG)
# Sample
zP_THUG = MSTHUG_adaptiveδ.sample()
```

### MS SwitchIntegrator with Adaptive $\delta$

```python
SETTINGS50_RWM_THEN_THUG = generate_setting(
    m=m, B=B, δ=δ, N=N, thinning=thinning, δ0=δ0, ϵ0=ϵ0,
    kernel_type=kernel_type, verbose=True, initialization='init_prior',
    burnin=burnin, maxiter=maxiter, ϵmin=1e-8, quantile_value=quantile_value, ϵprop_switch=1e-3,
    pmoved_switch=0.05, switch_strategy=switch_strategy, δmin=1e-3, δmax=δmax, ap_target=0.2)
```

```python
# Instantitate the algorithm
MSRWM_THUG_adaptiveδ = MSAdaptiveTolerancesAdaptiveδSwitchIntegrator(SETTINGS50_RWM_THEN_THUG)
# Sample
zP_RWM_THUG_adaptiveδ = MSRWM_THUG_adaptiveδ.sample()
```
