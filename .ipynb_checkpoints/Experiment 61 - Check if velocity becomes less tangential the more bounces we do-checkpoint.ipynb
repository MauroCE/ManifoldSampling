```python
import numpy as np
from numpy import zeros, eye, diag, exp, log
from numpy.linalg import norm, solve
from scipy.stats import multivariate_normal as MVN

import matplotlib.pyplot as plt

from tangential_hug_functions import HugTangential
from Manifolds.GeneralizedEllipse import GeneralizedEllipse
```

```python
# Manifold
μ       = zeros(2)
Σ       = diag([1.0, 5.0])
target  = MVN(μ, Σ)
z0      = -2.9513586307684885                # Specifies exactly which ellipse we are interested in (y*).
ellipse = GeneralizedEllipse(μ, Σ, exp(z0))
```

```python
# Functions for THUG
q = MVN(zeros(2), eye(2))
logprior      = lambda xi: MVN(zeros(2), eye(2)).logpdf(xi)


def log_epanechnikov_kernel(xi, epsilon, ystar):
    u = norm(target.logpdf(xi) - ystar)
    with np.errstate(divide='ignore'):
        return log((3*(1 - (u**2 / (epsilon**2))) / (4*epsilon)) * float(u <= epsilon))
```

```python
# Settings
ϵ = 0.01
T = 1.0
B = 5
N = 1
```

```python
logpi = lambda xi: logprior(xi) + log_epanechnikov_kernel(xi, ϵ, z0)
grad  = lambda xi: -solve(Σ, xi - μ)
```

```python
def integrator_initial(x0, v0s, α, B, T, grad_log_pi, normalized_output=False):
    """Tangential Hug integrator. Computes components of velocities along the tangent at the 
    initial point x0."""
    # Returns component along tangent and gradient
    components = np.zeros((B+2, 2))
    g0 = grad_log_pi(x0)
    g0 = g0 / norm(g0)
    v0 = v0s - α*g0*(g0@v0s)
    if not normalized_output:
        def comps(v):
            return norm(v - g0*(g0@v)), norm(g0@v)
    else:
        # In this case we return the proportion of the squared components over the total squared norm of v
        def comps(v):
            return norm(v - g0*(g0@v))**2 / norm(v)**2, norm(g0@v)**2 / norm(v)**2
    components[0, :] = comps(v0)
    v, x = v0, x0
    δ = T / B
    for _ in range(B):
        x = x + (δ/2)*v
        g = grad_log_pi(x)
        g = g / norm(g)
        v = v - 2*(v@g)*g
        components[_ + 1, :] = comps(v)
        x = x + (δ/2)*v
    g = grad_log_pi(x)
    g = g / norm(g)
    v = v + (α / (1-α))*g*(g@v)
    components[-1, :] = comps(v)
    return components

def integrator_current(x0, v0s, α, B, T, grad_log_pi, normalized_output=False):
    """Tangential Hug integrator. Computes components of velocities along the tangent at the 
    CURRENT POINT."""
    if not normalized_output:
        def comps(v, g):
            return norm(v - g*(g@v)), norm(g@v)
    else:
        def comps(v, g):
            assert isclose(norm(g), 1.0), "Norm should be one."
            vnorm_sq = norm(v)**2
            tangent_prop = norm(v - g*(g@v))**2 / vnorm_sq
            normal_prop  = norm(g@v)**2 / vnorm_sq
            assert isclose(tangent_prop + normal_prop, 1.0), "Should sum up to one."
            return tangent_prop, normal_prop
    components = zeros((B+2, 2))
    g0 = grad_log_pi(x0)
    g0 = g0 / norm(g0)
    v0 = v0s - α*g0*(g0@v0s)
    components[0, :] = comps(v0, g0)
    v, x = v0, x0
    δ = T / B
    for _ in range(B):
        x = x + (δ/2)*v
        g = grad_log_pi(x)
        g = g / norm(g)
        v = v - 2*(v@g)*g
        x = x + (δ/2)*v
        components[_+1, :] = comps(v, grad_log_pi(x) / norm(grad_log_pi(x)))
    g = grad_log_pi(x)
    g = g / norm(g)
    v = v + (α / (1-α))*g*(g@v)
    components[-1, :] = comps(v, g)
    return components
    

def function(n_runs, n_velocities, T, B, αlist, grad_log_pi, method='initial', norm_out=False):
    """Something."""
    if method == 'initial':
        integrator = integrator_initial
    elif method == 'current':
        integrator = integrator_current
    else:
        raise NotImplementedError("Not implemented.")
        
    n_alphas = len(αlist)
    COMPONENTS = np.zeros((n_runs, n_velocities, n_alphas, B + 2, 2))
    for run_ix in range(n_runs):
        
        # Sample a point
        x0 = ellipse.sample()
        
        # For each iteration in `n_velocities` sample a velocity and run THUG integrator
        for vel_ix in range(n_velocities):
            v = q.rvs()
            for α_ix, α in enumerate(αlist):
                COMPONENTS[run_ix, vel_ix, α_ix, :, :] = integrator(x0, v, α, B, T, grad_log_pi, normalized_output=norm_out)
    return COMPONENTS
```

# Initial

```python
n_runs = 30
n_vels = 10
Ti = 10.0
Bi = 100
αs = [0.0, 0.3, 0.5, 0.7, 0.9]

out = function(n_runs, n_vels, Ti, Bi, αs, grad, norm_out=True).mean(axis=1).mean(axis=0)[:, :-1, :]
```

```python
PLOT_UNTIL = Bi
fig, ax = plt.subplots(ncols=2, figsize=(10, 4))
for ix in range(2):
    for α_ix, α in enumerate(αs):
        ax[ix].plot(out[α_ix, :PLOT_UNTIL, ix], label=f"{α}")
ax[0].set_title("Tangent")
ax[1].set_title("Normal")
plt.suptitle("Directions at Initial Point")
plt.legend()
plt.show()
```

# Current

```python
Tc = Ti
Bc = Bi
```

```python
out_c = function(n_runs, n_vels, Tc, Bc, αs, grad, method='current', norm_out=True).mean(axis=1).mean(axis=0)[:, :-1, :]
```

```python
PLOT_UNTIL = Bc
fig, ax = plt.subplots(ncols=2, figsize=(10, 4), sharey=True)
for ix in range(2): 
    for α_ix, α in enumerate(αs):
        ax[ix].plot(out_c[α_ix, :PLOT_UNTIL, ix], label=f"{α}")
ax[0].set_title("Tangent")
ax[1].set_title("Normal")
plt.legend()
plt.suptitle("Directions at Current points.")
plt.show()
```
